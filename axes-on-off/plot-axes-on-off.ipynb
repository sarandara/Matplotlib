```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(77)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='gray')
```

```python
plt.imsave('plot-save-gray-1.jpg',arr,cmap='gray')
```

```python
np.random.seed(77)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='viridis')
plt.axis('off');
```

```python
plt.imsave('plot-save-viridis-1.jpg',arr,cmap='viridis')
```

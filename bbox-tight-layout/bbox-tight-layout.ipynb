```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(111)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='magma')
plt.savefig('plot-magma-1.jpg',cmap='magma')
```

```python
np.random.seed(111)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='magma')
plt.savefig('plot-magma-2.jpg',cmap='magma',bbox_inches='tight')
```

```python
np.random.seed(111)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='magma')
plt.axis('off');
plt.savefig('plot-magma-2-1.jpg',cmap='magma',bbox_inches='tight')
```

```python
np.random.seed(111)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='magma')
plt.tight_layout();
plt.savefig('plot-magma-3.jpg',cmap='magma')
```

```python
np.random.seed(111)
arr = np.random.randint(0,2,size=(256,256))
plt.imshow(arr,cmap='magma')
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0);
plt.savefig('plot-magma-3-1.jpg',cmap='magma')
```
